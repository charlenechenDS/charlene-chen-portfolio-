###Project Problem and Hypothesis

Domain Knowledge

    Video viewability has been a hot topic in ad tech industry. GroupM has developed its own metric to measure viewability called "GroupM In-view Rate" - stricter than industry-wide MRC IVR. In Moat reporting, we use the feature "Human, Audible & Fully On-Screen for Half of Duration (15 sec. cap) Rate" as GroupM IVR. According to Moat, the video viewability benchmark for 2016 Q1 is 22.1% for all programmatic media buy.

Problem Statement 

    This project is to determine which features (Advertiser vertical, domain/site, creative size, supply vendor, browser, non-human traffic rate, and small player size rate) will lead to higher (>=22.1%) GroupM IVR and what different roles they are playing (high coefficient vs. low coefficient) when it comes to higher GroupM IVR, based on Moat data from January-March 2016.

Hypothesis
    Advertiser vertical, domain/site, creative size, supply vendor, browser, non-human traffic rate, and small player size rate will allow us to predict if a video will have high viewability. (GroupM IVR>=22.1%)

Machine Learning Methodology 

    First of all, I will throw all data to Random Forrest to choose which features to look deeper. Then, use Lasso to calculate out each co-efficient for top 80% features from RF. To predict outcome - if a video will have high viewability (GroupM IVR>=22.1%), I will use Logistic Regression model to predict a binary value. (High viewability == 1, GroupM IVR>=22.1%; Low viewability == 0, GroupM IVR<22.1%)

Human, Audible & Fully On-Screen for Half of Duration (15 sec. cap) Rate
The percentage of impressions where the ad, for at least half the time (15 sec. cap), was audible while the ad surface was 100% on-screen and was delivered to a human. For players 300,000 square pixels or more, only 80% of pixels are required to be on-screen.

Percentage of impressions where the width of the player was less than 400px.

Below the Fold Rate
Percentage of impressions where more than half of the ad loaded below the fold. The fold is defined as 750px.

Out of Focus Rate
Percentage of impressions where the tab containing the ad was never brought into foreground.
